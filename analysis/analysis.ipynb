```python
import pandas as pd
import json
```

```python
output_json = "sample.json"

# try:
#   with open(output_json) as infile:
#     job_desc_id = json.load(infile)
# except Exception as e:
#   print(e)

df = pd.read_json(output_json)
```

```python
df = df.T
df = df.reset_index()
df = df.rename(columns={"index": "hash_id", "job_title": "job_title", "job_desc": "job_description"})
```

```python
df.head()
```

```python
df.shape
```

```python
clean_df = df[df['job_description'].notna()]
clean_df = clean_df.replace('\n',' ', regex=True)
```

```python
clean_df.head()
```

```python
clean_df.shape
```

```python
job_description = clean_df['job_description'].tolist()
```

```python
print(type(job_description))
print(len(job_description))
print(job_description[:3])
```

Keyword extraction - keyBERT

```python
from keybert import KeyBERT

results_keybert = []
kw_model = KeyBERT()

for jd in job_description:
    results_keybert.append(kw_model.extract_keywords(jd, keyphrase_ngram_range=(1, 3), 
                          use_maxsum=True, nr_candidates=20, top_n=5))
```

```python
results_keybert[:3]
```

```python
results_keybert_simplified = []
for _listings in results_keybert:
    _inner_list = []
    for _extracted_keywords in _listings:
        _inner_list.append(_extracted_keywords[0])
    results_keybert_simplified.append(_inner_list)
```

```python
results_keybert_simplified[0]
```

```python
formatted = [" ".join(x) for x in results_keybert_simplified]
```

```python
formatted
```

```python
len(formatted)
```

```python
formatted_series = pd.Series(formatted)
```

```python
# Create a deep copy of the DataFrame
added_df = clean_df.copy(deep=True)
```

```python
added_df['extracted'] = formatted_series
```

```python
added_df
```

```python
added_df.to_csv('extracted_df.csv')  
```
